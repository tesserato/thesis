```python
import sympy as sp

x = sp.IndexedBase("x", dim=1, real=True, positive=True)
y = sp.IndexedBase("y", dim=1, real=True, positive=True)

s, xc, yc = sp.symbols("s x_c y_c", real=True, positive=True)
i, n = sp.symbols("i n", integer=True, positive=True)
#m0 = sp.symbols("m_0", real=True)

def m(i):
    return (s * y[i + 1] - s * y[i]) / (x[i + 1] - x[i])

m(1)
```

```python
def get_circle(x0, y0, x1, y1, r):
    '''returns center of circle that passes through two points'''

    radsq = r * r
    q = sp.sqrt((x1 - x0)**2 + (y1 - y0)**2)

    x3 = (x0 + x1) / 2
    y3 = (y0 + y1) / 2

    xc = x3 + sp.sqrt(radsq - (q / 2)**2) * ((y0 - y1) / q)
    yc = y3 + sp.sqrt(radsq - (q / 2)**2) * ((x1 - x0) / q)

    return xc, yc

xc, yc = get_circle(x[i], s * y[i], x[i+1], s * y[i+1], ri)
xc
```
