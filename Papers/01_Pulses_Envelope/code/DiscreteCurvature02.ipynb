```python
import sympy as sp

x0, x1, y0, y1, y2, x, y, m, r = sp.symbols("x_0 x_1 y_0 y_1 y_2 x y m r", real=True, positive=True)

eq1 = sp.Eq( (y - y0), (-1/m) * (x - x0) )
eq1
```

```python
m2 = (y1 - y0) / (x1 - x0)
m2
```

```python
eq2 = sp.Eq( (y - y1), (-1/m2) * (x - x1) )
eq2
```

```python
yc = sp.solve(eq1, y)[0]
print(yc)
```

```python
xc = sp.solve(eq2.subs(y, yc), x)[0]
print(xc)
```

```python
r1 = sp.sqrt((x0 - xc)**2 + (y0 - yc)**2)
r1
```

```python
r1.simplify()
```

```python
r2 = sp.sqrt((x1 - xc)**2 + (y1 - yc)**2)
r2
```

```python
r2.expand() == r1.expand()
```
