```python
import sympy as sp

s, x, y = sp.symbols("s x y", real=True, positive=True)
m_ = sp.symbols("\overline{m}", real=True, positive=True)

m = y / x
```

```python
theta = sp.atan( (m - m_) / (1 + m * m_) ).expand().factor(deep=True).simplify(trig=True)
theta
```

```python
r = ( x / sp.sin(theta) ).expand(trig=True).simplify(trig=True).factor(deep=True)
r
```

```python
r.subs(m_, 0)
```
