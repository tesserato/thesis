```python
import sympy as sp

x, y, x0, x1, y0, y1, a, b, c, r, xc, yc = sp.symbols("x y x_0 x_1 y_0 y_1 a b c r x_c y_c", real=True)
f = sp.Function("f", size=1)(x)
```

```python
base = sp.sqrt( (x1 - x0)**2 + (y1 - y0)**2 )
base
```

```python
alpha = (sp.atan( (y1 - y0) / (x1 - x0) )).simplify(trig=True)
alpha
```

```python
rho = sp.acos((y0 - y1)/(2*(x0 - x1))) + sp.atan((-y0 + y1)/(-x0 + x1)) + 3*sp.pi/2
print(rho)
```

```python
rho.expand(trig=True)#.simplify(trig=True)
```

```python
theta = (sp.pi/2 - (alpha - rho)).simplify(trig=True)
theta
```

```python
r = (base / 2) / sp.cos(theta) # cos = (b/2) / r
r
```

```python
eq = sp.Eq(2 * sp.cos(b), (y1-y0)/(x1-x0))
eq
```

```python
sp.solve(eq, b)[0]
```

```python
'''rho = theta + alpha - pi/2'''
rho = sp.solve(eq, b)[0] + sp.atan( (y1 - y0) / (x1 - x0) ) - sp.pi/2
rho
```
