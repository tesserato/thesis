```python
import sympy as sp

a, b, c, x = sp.symbols("a b c x", real=True, positive=True)

y = a * sp.tanh(b / a).expand(trig=True)
y
```

```python
eq = sp.Eq(c/a, y/a)
eq
```

```python
(y/a).subs(x, b).expand(trig=True).simplify(trig=True)
```

```python
eq = sp.Eq(sp.log(sp.exp(x/a)**2**2-1).subs(x, b), sp.log(c/a)).expand()#.simplify(trig=True)
eq
```

```python
eq2 = sp.Eq(sp.log(sp.exp(x/a).subs(x, b)**2**2), sp.log(c/a+1).expand(trig=True))
eq2
```

```python
aa = sp.solve(eq2, a)
aa
```

```python
bb = sp.solve(eq, b)[0].expand(trig=True).simplify(trig=True)
bb
```

```python
aa.subs(b, bb).simplify()
```

```python
eq2 = sp.Eq(aa.subs(b, bb).simplify(), b).rewrite(sp.exp).expand().simplify()
eq2
```
