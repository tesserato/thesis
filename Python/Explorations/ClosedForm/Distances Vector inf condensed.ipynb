```python
import sympy as sp

n, t = sp.symbols("n t", integer=True, positive=True)
x0, y0 = sp.symbols("x_0 y_0", real = True, positive=True)
k, s = sp.symbols("k_t s_t", integer=True, positive=True)
W = sp.IndexedBase("W")

x_vec = (2 * sp.pi * n**2) / (n**2 + 4 * sp.pi**2 * t**2)     # p
y_vec = (4 * sp.pi**2 * n * t) / (n**2 + 4 * sp.pi**2 * t**2) # f
m_vec = sp.sqrt(x_vec**2 + y_vec**2)

U=sp.Matrix([x_vec, y_vec]) / m_vec
P=sp.Matrix([x0, y0])
O=sp.Matrix([s * sp.pi, 0])

d=sp.sqrt((P-O).T @ U@U.T @(P-O))[0, 0].simplify()
d
```
