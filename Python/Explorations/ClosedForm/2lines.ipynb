```python
import sympy as sp

R = sp.Matrix([[0, -1], [1, 0]])

xl = sp.IndexedBase("xl")
yl = sp.IndexedBase("yl")
def P(i):
    return sp.Matrix([xl[i], yl[i]])

xv = sp.IndexedBase("xv")
yv = sp.IndexedBase("yv")
def V(i):
    return sp.Matrix([xv[i], yv[i]])

#xn = sp.IndexedBase("xn")
#yn = sp.IndexedBase("yn")
def N(i):
    Vi = V(i)
    return Vi / sp.sqrt(Vi.T @ Vi)[0, 0]

#xu = sp.IndexedBase("xu")
#yu = sp.IndexedBase("yu")
def U(i):
    return R @ N(i)

i, n = sp.symbols("i n", integer=True, positive=True)

sp.Sum(N(i) @ N(i).T, (i, 0, n))**(-1)
```
