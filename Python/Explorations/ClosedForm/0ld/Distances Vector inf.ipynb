```python
import sympy as sp

n, t = sp.symbols("n t", integer=True, positive=True)
x0, y0 = sp.symbols("x_0 y_0")
k = sp.IndexedBase("k_t", integer=True)
W = sp.IndexedBase("W")

x_vec = (2 * sp.pi * n**2) / (n**2 + 4 * sp.pi**2 * t**2)     # p
y_vec = (4 * sp.pi**2 * n * t) / (n**2 + 4 * sp.pi**2 * t**2) # f
m_vec = sp.sqrt(x_vec**2 + y_vec**2)

U=sp.Matrix([x_vec,y_vec]) / m_vec
P=sp.Matrix([x0, y0])
O_pos=sp.Matrix([0, 0])
O_neg=sp.Matrix([sp.pi, 0])

d_pos=sp.sqrt((P-O_pos).T @ U@U.T @(P-O_pos))[0, 0].simplify()
d_pos
```

```python
d_neg=sp.sqrt((P-O_neg).T @ U@U.T @(P-O_neg))[0, 0].simplify()
d_neg
```

```python
m_vec.simplify()
```

```python
r_pos = (d_pos - m_vec * k).expand().simplify()
r_pos
```

```python
r_neg = (d_neg - m_vec * k).expand().simplify()
r_neg
```

```python
wr_pos = W[t] * r_pos
wr_pos
```

```python
wr_neg = W[t] * r_neg
wr_neg
```

```python
dwrdx0_pos = sp.diff(wr_pos, x0).simplify()
dwrdx0_pos
```

```python
dwrdy0_pos = sp.diff(wr_pos, y0).simplify()
dwrdy0_pos
```

```python
eq_pos = (n * x0 + 2 * sp.pi * t * y0) * (n**2 + 4 * sp.pi**2 * t**2) * W[t]
eq_pos.expand().simplify()
```

```python
dwrdx0_neg = sp.diff(wr_neg, x0).simplify()
dwrdx0_neg
```

```python
dwrdy0_neg = sp.diff(wr_neg, y0).simplify()
dwrdy0_neg
```

```python
eq_neg = (n * x0 + 2 * sp.pi * t * y0) * (n**2 + 4 * sp.pi**2 * t**2) * W[t]
eq_neg.expand().simplify()
```
