```python
import sympy as sp

n, t, p, fm = sp.symbols("n t p f_m")
W = sp.IndexedBase("W")

f0 = n / t
f2pi = 0

b=f0
a=(f2pi-f0) / (2 * sp.pi)
f = (a * p + b)
A = sp.Sum((f * W[t]-fm)**2,(t,0,n-1))
sp.diff(A, fm)
```
