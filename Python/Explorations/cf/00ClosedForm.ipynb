```python
import sympy as sp

R = sp.Matrix([[0, -1], [1, 0]])

i, n = sp.symbols("i n", integer=True, positive=True)
x0, y0, g = sp.symbols("x_0 y_0 \gamma", real=True, positive=True)

xl = sp.IndexedBase("xl", real=True)
yl = sp.IndexedBase("yl", real=True)
W = sp.IndexedBase("W", shape=1, real=True, positive=True)

def P(i):
    return sp.Matrix([xl[i], yl[i]])

xv = sp.IndexedBase("xv")
yv = sp.IndexedBase("yv")
def V(i):
    return sp.Matrix([xv[i], yv[i]])

def N(i):
    Vi = V(i)
    return Vi / sp.sqrt(Vi.T @ Vi)[0, 0]

def U(i):
    return R @ N(i)

def line(i):
    return P(i) + g * U(i)

X0 = sp.Matrix([x0, y0])

def d2(i):
    return W[i] * (X0 - P(i)).T @ (N(i) @ N(i).T) @ (X0 - P(i))

S = sp.Sum(d2(i), (i, 0, n-1))
S
```

```python
dSdx = sp.diff(S, X0).simplify()
dSdx
```
