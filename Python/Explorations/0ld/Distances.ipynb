```python
import sympy as sp

n, t, p, f0, p0 = sp.symbols("n t p f0 p0")
W = sp.IndexedBase("W")

f1 = n / t
f2 = 0
p1 = 0
p2 = 2 * sp.pi

d2 = ((f2 - f1) * p0 - (p2 - p1) * f0 + p2 * f1 - p1 * f2)**2 / ((f2-f1)**2+(p2-p1)**2)
A = sp.Sum(d2 * W[t],(t,1,n-1))
sp.diff(A, f0)#.expand().simplify()
```

```python
import plotly.graph_objects as go
import numpy as np


n = 200
X = np.arange(n)
f = 3
p = np.pi / 2
W = np.cos(p + 2 * np.pi * f * X / n)

Q = np.zeros((n-1, 2))
Y = np.zeros((n-1))

for t in range(1, n):
    #den = 4 * np.pi**2  + (n/t)**2
    Q[t-1, 1] = W[t] *  2 * np.pi
    Q[t-1, 0] = W[t] * (n/t)
    Y[t-1] = W[t] * 2 * np.pi * (n/t)
A = np.linalg.inv(Q.T @ Q) @ Q.T @ Y

np.round(A, 3)
```
