$$ \mathcal{I}(f,\phi) = \sum_{x=0}^{n-1}  W[x] \cos \left( \phi + 2 \pi f \frac{x}{n} \right) $$

```python
import sympy as sp

n, x, k, s = sp.symbols("n x k s", integer=True, positive=True)
p0, f0, p, f, w = sp.symbols("\phi_0 f_0 \phi f \omega", real=True, positive=True)
W = sp.Function("W", real=True)

def cos(x):
    return sp.Sum((-1)**k * x**(2*k) / sp.factorial(2 * k), (k, 0, 3))

def a(x):
    return W(x) * sp.cos(p + w * x) # w = 2 * sp.pi * f / n

I = sp.integrate(a(x), (x,0,n-1)).doit()

I
```

```python
I.expand(trig=True).simplify()
```

```python
sp.cos(x).series(x,s,5)
```

```python
dIdf = sp.diff(I, w).expand(trig=True)
dIdf
```

```python
u  = x
dv = W(x)

v  = dv.integrate((x, 0, n-1)) 
du = u.diff(x)

(u * dv).integrate((x, 0, n-1))
```

```python
(u * s - (s * du).integrate((x, 0, n-1))).subs(s, v).simplify()
```

```python
u  = sp.cos(w * x / n)
dv = x*W(x)

v  = W(x).integrate((x, 0, n-1)) * (x+1-n)
du = u.diff(x)

(u * dv).integrate((x, 0, n-1))
```

```python
(u * v - (v * du).integrate((x, 0, n-1))).simplify()
```
