$$ \mathcal{I}(f,\phi) = \sum_{x=0}^{n-1}  W[x] \cos \left( \phi + 2 \pi f \frac{x}{n} \right) $$

```python
import sympy as sp

n, x, k, s = sp.symbols("n x k s", integer=True, positive=True)
p0, f0, p, f, w = sp.symbols("\phi_0 f_0 \phi f \omega", real=True, positive=True)
W = sp.Function("W", real=True)(x)

def term(k):
    return (-1)**k * x**(2*k) / sp.factorial(2 * k)

def cos(x):
    return sp.Sum(W * term(k), (k, 0, "oo"))

def a(x):
    return  cos(p + x * 2 * sp.pi * f / n) # w = 2 * sp.pi * f / n


I = sp.Sum(a(x), (x,0,n-1)).expand(trig=True).doit()#.expand(trig=True).simplify()
I
```
