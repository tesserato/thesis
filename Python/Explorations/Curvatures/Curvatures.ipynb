```python
import sympy as sp

x, x0, x1, a, b, c = sp.symbols("x x_0 x_1 a b c", real=True)
f = sp.Function("f", size=1)(x)
y = a * x**2 + b * x + c
y
```

```python
'''curvature:'''
f.diff(x,x) / ( (1 + f.diff(x)**2)**sp.Rational(3 / 2) )
```

$$ \kappa = \frac{f''(x)}{(f'(x)^2 + 1)^{\frac{3}{2}}} $$

```python
k = y.diff(x,x) / ( (1 + y.diff(x)**2)**sp.Rational(3 / 2) )
k
```

```python
x_max_k = sp.solve(k.diff(x), x)[0]
x_max_k
```

```python
k.subs(x, x_max_k)
```

```python
sp.integrate(k, (x, x0, x1)).simplify() # (2a2)/(a1^2+4a1a2x+4a2^2x^2+1)^(3/2)
```

$$ \dfrac{2 a x + b}{\sqrt{\left(2 a x + b \right)^2 + 1}} $$

```python
t1 = 2 * a * x1 + b
t0 = 2 * a * x0 + b
c = t1 / sp.sqrt(t1**2 + 1) - t0 / sp.sqrt(t0**2 + 1)
c
```
