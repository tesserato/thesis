```python
import sympy as sp

x, x0, x1, x2 = sp.symbols("x x_0 x_1 x_2", real=True, positive=True)
y, y0, y1, y2 = sp.symbols("y y_0 y_1 y_2", real=True)
a, b, c, d, e, f = sp.symbols("a b c d e f", real=True)

y = a * x**2 + b * x + c
y
```

```python
e = (x - x0)**2 + (y - y0)**2
e.simplify()
```

```python
e.diff(x).expand()
```

```python
c = (y.diff(x,x) / ( (1 + y.diff(x)**2)**sp.Rational(3 / 2) )).expand().factor(deep=True)
print(c)
```

```python
c = (y.diff(x,x) / ( sp.sqrt(1 + y.diff(x)**2)**(3) ) ).expand().simplify()#.factor(deep=True)
c
```
