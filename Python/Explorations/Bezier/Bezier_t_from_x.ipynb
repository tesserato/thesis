```python
import sympy as sp

x, x0, x1, x2, xl1, xl2, xl3, t, tt = sp.symbols("x x_0 x_1 x_2 xl_1 xl_2 xl_3 t tt", real=True, positive=True)
y, y0, y1, y2, yl1, yl2, yl3 = sp.symbols("y y_0 y_1 y_2 yl_1 yl_2 yl_3", real=True)

eq1 = (x - xl1) / (xl2 - xl1)
eq1
```

```python
eq2 = (xl1 - x0) / (x1 - x0)
eq2
```

```python
eq3 = (xl2 - x1) / (x2 - x1)
eq3
```

```python
xl2_ = sp.solve(sp.Eq(eq2, eq3), xl2)[0]
xl2_
```
