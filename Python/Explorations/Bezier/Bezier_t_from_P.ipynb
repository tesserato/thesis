```python
import sympy as sp

x, x0, x1, x2, xl1, xl2, t = sp.symbols("x x_0 x_1 x_2 xl_1 xl_2 t", real=True, positive=True)
y, y0, y1, y2, yl1, yl2, d = sp.symbols("y y_0 y_1 y_2 yl_1 yl_2 d", real=True)

P = sp.geometry.Point(x, y)
P0 = sp.geometry.Point(x0, y0)
P1 = sp.geometry.Point(x1, y1)
P2 = sp.geometry.Point(x2, y2)
Pl1 = sp.geometry.Point(xl1, yl1)
Pl2 = sp.geometry.Point(xl2, yl2)

```

```python
Pl1 = P0 + (P1 - P0) * t
Pl1
```

```python
Pl2 = P1 + (P2 - P1) * t
Pl2
```

```python
PP = Pl1 + (Pl2 - Pl1) * t
PP
```

```python
eqx = sp.Eq(PP.coordinates[0], P.coordinates[0])
eqx
```

```python
eqy = sp.Eq(PP.coordinates[1], P.coordinates[1])
eqy
```

```python
PP.coordinates[1].expand()
```

```python
tx = sp.solve(eqx, t)
tx[0]
```
