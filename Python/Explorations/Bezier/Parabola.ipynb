```python
import sympy as sp

x, x0, x1, x2 = sp.symbols("x x_0 x_1 x_2", real=True, positive=True)
y, y0, y1, y2 = sp.symbols("y y_0 y_1 y_2", real=True)
a, b, c, d, e, f = sp.symbols("a b c d e f", real=True)

y = a * x**2 + b * x + c
y
```

```python
eq1 = sp.Eq(y.diff(x).subs(x, x0), (y1 - y0) / (x1 - x0))
eq1
```

```python
eq2 = sp.Eq(y.diff(x).subs(x, x2), (y2 - y1) / (x2 - x1))
eq2
```

```python
sol = sp.solve([eq1, eq2], (a, b))
```
