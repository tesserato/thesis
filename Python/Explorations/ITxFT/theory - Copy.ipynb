```python
import sympy as sp

n, t1, t2 = sp.symbols("n t_1 t_2", integer=True, positive=True)
f, p, p1, p2 = sp.symbols("f p p1 p2", real = True, positive=True)
k, s = sp.symbols("k_t s_t", integer=True, positive=True)
W = sp.IndexedBase("W", real = True, shape=1)

FP1 = W[t1] * sp.cos(p + 2 * sp.pi * f * t1 / n)
FP2 = W[t2] * sp.cos(p + 2 * sp.pi * f * t2 / n)

eq = FP1 + FP2

eq
```

```python
eq.expand(trig=True).trigsimp()#.subs(sp.cos(p1 + 2 * sp.pi * f * t1 / n), 1).subs(sp.cos(p2 + 2 * sp.pi * f * t2 / n), 1)
```

```python
sol = sp.solve(sp.diff(eq, p).expand(trig=True).trigsimp(), p)
```
