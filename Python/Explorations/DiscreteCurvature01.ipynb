```python
import sympy as sp

x0, x1, x2, y0, y1, y2, x, y, m, r = sp.symbols("x_0 x_1 x_2 y_0 y_1 y_2 x y m r", real=True, positive=True)

eq1 = sp.Eq(y-(y0 + y1) / 2, (-1/m) * (x-(x0 + x1) / 2))
eq1
```

```python
eq2 = sp.Eq(sp.sqrt((x-x0)**2 + (y-y0)**2), r)
eq2
```

```python
yy = sp.solve(eq1, y)[0]
print(yy)
```

```python
xx = sp.solve(eq2.subs(y, yy), x)
xx[0]
```
