```python
import sympy as sp

n, t = sp.symbols("n t", integer=True, positive=True)
x0, y0 = sp.symbols("x_0 y_0", real = True, positive=True)
k, s = sp.symbols("k_t s_t", integer=True, positive=True)
W = sp.IndexedBase("W", real = True)

Vx = (2 * sp.pi * n**2) / (n**2 + 4 * sp.pi**2 * t**2)     # p
Vy = (4 * sp.pi**2 * n * t) / (n**2 + 4 * sp.pi**2 * t**2) # f
m = (sp.sqrt(Vx**2 + Vy**2)).simplify()

V = sp.Matrix([Vx, Vy]) # == sp.sqrt(V.T @ V)

U = V / Vm
P=sp.Matrix([x0, y0])
O=sp.Matrix([0, 0])

d=sp.sqrt((P-O).T @ U@U.T @(P-O))[0, 0].simplify()

r_pos = ((d - k * m)).expand().simplify()
r_pos
```

```python
r_neg = ((d - k * m + m/2)).expand().simplify()
r_neg
```
