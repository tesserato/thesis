```python
import plotly.graph_objects as go
import numpy as np
import sympy as sp



n = 200
X = np.arange(n)
f = 2
p = np.pi / 2
W = np.cos(p + 2 * np.pi * f * X / n)

Q = np.zeros((n-1, 3))
Y = np.zeros((n-1))

N = np.zeros((1, 2)).T
p = np.array([[2 * np.pi], [0]])
Sn = np.zeros((2, 2))
Sp = np.zeros((2, 1))

for t in range(n):
    p_vec = (2 * np.pi * n**2) / (n**2 + 4 * np.pi**2 * t**2)
    f_vec = (4 * np.pi**2 * n * t) / (n**2 + 4 * np.pi**2 * t**2)
    l = np.sqrt(p_vec**2 + f_vec**2)
    N[0, 0] = p_vec/l
    N[1, 0] = f_vec/l
    nnt = N @ N.T #* np.abs(W[t])
    Sn += nnt
    #p = np.array([[2 * np.pi], [0]]) if W[t]>0 else np.array([[3 * np.pi], [0]])
    Sp += nnt @ p

A = np.linalg.inv(Sn) @ Sp

np.round(A, 3)#[0] % (2*np.pi)
```
