```python
import sympy as sp

x, x0, x1, x2, xl1, xl2, xl3, t = sp.symbols("x x_0 x_1 x_2 xl_1 xl_2 xl_3 t", real=True, positive=True)
y, y0, y1, y2, yl1, yl2, yl3 = sp.symbols("y y_0 y_1 y_2 yl_1 yl_2 yl_3", real=True)

eq1 = (x - xl1) / (xl2 - xl1)
eq1
```

```python
eq2 = (xl1 - x0) / (x1 - x0)
eq2
```

```python
eq3 = (xl2 - x1) / (x2 - x1)
eq3
```

```python
xl2_ = sp.solve(sp.Eq(eq2, eq3), xl2)[0]
xl2_
```

```python
xl1_ = sp.solve(sp.Eq(eq1.subs(xl2, xl2_).factor(), eq2), xl1)
xl1_[0]
```

```python
Xp = [1, 3, 6]
Yp = [3, 2, 4]

f = yl3.subs(x0, Xp[0]).subs(x1, Xp[1]).subs(x2, Xp[2]).subs(y1, Yp[1]).subs(y0, Yp[0]).subs(y2, Yp[2])
f
```

```python
X = []
Y = []
for i in range(min(Xp), max(Xp) + 1):
    X.append(i)
    Y.append(f.subs(x, i).evalf())
Y
```
