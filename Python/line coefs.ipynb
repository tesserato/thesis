```python
import sympy as sp
x = sp.IndexedBase("x", real=True)
y = sp.IndexedBase("y", real=True)
a, b = sp.symbols("a b", real=True)

eq1 = sp.Eq(y[1], a * x[1] + b)
eq1
```

```python
eq2 = sp.Eq(y[2], a * x[2] + b)
eq2
```

```python
sp.solve([eq1, eq2], a, b)[a]
```

```python
sp.solve([eq1, eq2], a, b)[b]
```

```python
k1, k2, T = sp.symbols("k_1 k_2 T", real=True)
```

```python
((k1 + 1) * k2 - k1 * ( k2 + T)).simplify()
```
