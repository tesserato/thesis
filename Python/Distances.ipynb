```python
import sympy as sp

n, t, p, f0, p0 = sp.symbols("n t p f0 p0")
W = sp.IndexedBase("W")

f1 = n / t + 1
f2 = 0
p1 = 0
p2 = 2 * sp.pi

d2 = ((f2 - f1) * p0 - (p2 - p1) * f0 + p2 * f1 - p1 * f2)**2 / ((f2-f1)**2+(p2-p1)**2)
A=sp.Sum(d2 * W[t],(t,1,n-1))
sp.diff(A, f0)
```

```python
sp.diff(A, f0).subs(n, 4).doit()
```
