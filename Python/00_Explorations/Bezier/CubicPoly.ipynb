```python
import sympy as sp

x, x0, x1, x2 = sp.symbols("x x_0 x_1 x_2", real=True, positive=True)
y, y0, y1, y2 = sp.symbols("y y_0 y_1 y_2", real=True)
a, b, c, d, e, f = sp.symbols("a b c d e f", real=True)

y = a * x**3 + b * x**2 + c * x + d
y
```

```python
eq1 = sp.Eq(y.subs(x, x0), y0)
eq1
```

```python
eq2 = sp.Eq(y.subs(x, x2), y2)
eq2
```

```python
sol1 = sp.solve([eq1, eq2], (c, d))
sol1
```

```python
print(sol1[c].factor())
```

```python
eq3 = sp.Eq(y.diff(x).subs(x, x0), (y1 - y0) / (x1 - x0))
eq3
```

```python
eq4 = sp.Eq(y.diff(x).subs(x, x2), (y2 - y1) / (x2 - x1))
eq4
```

```python
sol = sp.solve([eq1, eq2, eq3, eq4], (a, b, c, d))
```

```python
sol[a].factor()
```

```python
sol[b].factor(deep=True)
```
