```python
import sympy as sp

'''fit free polynomials and then glue them a posteriori'''

x = sp.symbols("x", real=True, positive=True)
a = sp.IndexedBase("a", real=True)
b = sp.IndexedBase("b", real=True)
u = sp.IndexedBase("u", real=True)
v = sp.IndexedBase("v", real=True)

def poly(a, k=3):
    p = a[0]
    for i in range(1, k+1):
        p += a[i] * x**i
    return p

A, B, U, V = poly(a), poly(b), poly(u), poly(v)
```

```python
e = (A - B)**2
e
```

```python
sp.solve(e.diff(x), x)[0]
```
