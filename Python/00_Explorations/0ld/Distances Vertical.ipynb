```python
import plotly.graph_objects as go
import numpy as np


n = 20000
X = np.arange(n)
f = 6
p = 2 * np.pi
W = np.cos(p + 2 * np.pi * f * X / n)

Q = np.zeros((n-1, 2))
Y = np.zeros((n-1))

for t in range(1, n):  
    Q[t-1, 0] =  2 * abs(W[t])
    Q[t-1, 1] = (n / (np.pi * t)) * abs(W[t])
    Y[t-1] = (n / t) * abs(W[t]) if W[t] < 0  else 0

A = np.linalg.inv(Q.T @ Q) @ Q.T @ Y

A
```
