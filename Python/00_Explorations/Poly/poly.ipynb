$$ \mathcal{I}(f,\phi) = \sum_{x=0}^{n-1}  W[x] \cos \left( \phi + 2 \pi f \frac{x}{n} \right) $$

```python
import sympy as sp

n, x, k, s, x0, x1 = sp.symbols("n x k s_x x_0 x_1", integer=True, positive=True)
d, c, p, f, w = sp.symbols("d c \phi f \omega", real=True, positive=True)
W = sp.Function("W", real=True, size=1)(x)
v = sp.Function("v", real=True, size=1)(f, x) # k pi

def cos1(x):
    return sp.Sum(c*(-1)**k / (1 + d*(x - k * sp.pi - s)**2), (k, -sp.oo, sp.oo))

def cos2(x):
    return c/(d*(x-s)**2+1)+sp.Sum(c*(-1)**k / (1 + d*(x - k * sp.pi - s)**2) + c*(-1)**k / (1 +d*(x + k * sp.pi-s)**2), (k, 1, sp.oo))

def cos3(x):
    s = (x // (2 * sp.pi)) * 2 * sp.pi
    return sp.Sum(c*(-1)**k / (1 + d*(x - k * sp.pi - s)**2), (k, -10, 10))

def a(x):
    return W * cos2(p + (2 * sp.pi * f / n) * x) # w = (2 * sp.pi * f / n)

I = sp.Sum(a(x), (x,0,n-1))

I
```

```python
denominator = n**2 + 4 * sp.pi**2 * x**2
Vp = (2 * sp.pi * n**2) / denominator
Vf = (4 * sp.pi**2 * n * x) / denominator
Vm = sp.sqrt(Vp**2 + Vf**2).simplify()

V = sp.Matrix([Vp, Vf]) # == sp.sqrt(V.T @ V)

U = V / Vm
P = sp.Matrix([p0, f0])
O = sp.Matrix([s * sp.pi, 2 * sp.pi * k]) # s=0 if w[t] > 0, 1 cc

d = sp.sqrt((P-O).T @ U@U.T @(P-O))[0, 0].simplify()

l = sp.Eq(n * k / x - n * p / (2 * sp.pi * x) + n * s / (2 * x), f)
l
```
