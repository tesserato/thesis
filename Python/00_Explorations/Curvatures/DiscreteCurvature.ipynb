```python
import sympy as sp

x0, x1, x2, y0, y1, y2 = sp.symbols("x_0 x_1 x_2 y_0 y_1 y_2", real=True)

v1 = sp.Matrix([x1 - (x0 + x1) / 2, y1 - (y0 + y1) / 2]).T
v2 = sp.Matrix([x2 - (x1 + x2) / 2, y2 - (y1 + y2) / 2]).T
```

```python
x, y, m, r = sp.symbols("x y m r", real=True)
sp.Eq(y-(y0 + y1) / 2, (-1/m) * (x-(x0 + x1) / 2))
```

```python
sp.Eq(sp.sqrt((x-x0)**2 + (y-y0)**2), r)
```

```python
theta = sp.acos( v1.dot(v2) / (v1.norm() * v2.norm()) )
theta
```

```python
k = theta / (v1.norm() + v2.norm())
k
```

```python
m1 = ( (y1 - (y1 + y0) / 2) / (x1 - (x1 + x0) / 2) ).simplify()
m1
```

```python
m2 = ( (y2 - (y2 + y1) / 2) / (x2 - (x2 + x1) / 2) ).simplify()
m2
```

```python
k_ = sp.atan( (m2 - m1) / (1 + m1 * m2) ).simplify()
k_
```
