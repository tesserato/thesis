```python
import sympy as sp

t, n = sp.symbols("t n", integer=True, positive=True)

def V(t):
    denominator = (n**2 + 4 * sp.pi**2 * t**2)
    return sp.Matrix([2 * sp.pi * n**2 / denominator, 4 * sp.pi**2 * n * t / denominator])


V(t)
```

```python
V(t)[1] / V(t)[0]
```

```python
q = sp.IndexedBase("q", shape=1, real=True)
def U(t):
    return q[t] * sp.Matrix([1, 2 * sp.pi * t / n])

U(t)
```
