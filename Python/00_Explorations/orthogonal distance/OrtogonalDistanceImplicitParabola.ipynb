```python
import sympy as sp

x, x0, x1, x2 = sp.symbols("x x_0 x_1 x_2", real=True, positive=True)
y, y0, y1, y2 = sp.symbols("y y_0 y_1 y_2", real=True)
a, b, c, d, e, f = sp.symbols("a b c d e f", real=True)

p = a * x**2 + b * x * y + c * y**2 + d * x + e * y + f # == 0, b**2==4*a*c
sp.Eq(p, 0)
```

```python
y = sp.solve(p, y)[1].simplify()
y
```

```python
e = (x - x0)**2 + (y - y0)**2
e.expand().factor()
```
